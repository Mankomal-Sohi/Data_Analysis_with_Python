What if we want to return the total number of routes each bus cover.

We use value_counts function that returns a Series containing counts of unique values. In other words, it counts how many times each unique value appears in the 'Route Number' column.

```python
route_counts=data['Route Number'].value_counts() 
```

```python
route_counts
```

```python
route_counts[:10]  # it will print the top 10 rows
```

```python
# Now we will plot the graph which displays the top 10 most used bus routes

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
route_counts[:10].plot(kind='bar', color='red')

plt.title("Top 10 most used bus routes")
plt.xlabel("Route Number")
plt.ylabel("Total routes")

plt.show()
```

```python
# now we will print how many route number goes to each destination.

routes=data.groupby("Route Destination")['Route Number'].count()
```

```python
routes
```

```python
route=routes.nlargest(10) # Top 10 destinations where most buses passes by
```

```python
route
```

```python
# plot the graph that comapre the total bus routes going to each destination 

plt.figure(figsize=(8,6))
route.plot(kind='barh', color='blue')

plt.title("Number of buses to each destination")
plt.xlabel("Route Destination")
plt.ylabel("Total buses")

plt.show()
```

# DATASET 2

```python
# Firstly, we will retrieve data from the website

from urllib.request import urlretrieve
URL="https://data.seattle.gov/resource/65db-xm6k.csv"

urlretrieve(URL,'fremont.csv')
```

```python
data2=pd.read_csv('fremont.csv',parse_dates =["date"], index_col ="date")
```

```python
data2.columns = ['Total', 'South', 'North']  # name the coulumns 
```

```python
data2.head(10)  # prints top 10 rows
```

```python
data2.describe()
```

```python
# plot the total numbver of cycles passes by each hour
%matplotlib inline

data2.plot()
plt.ylabel('Hourly Bicycle Count');
```

```python
#Print mean of total number of cycles passed by per week in the north and south part

Weekly_resampled_data = data2.resample('W').mean() 
```

```python
Weekly_resampled_data
```

```python
# Now we will plot the line graph that shows the trend how many cycles passes the orth and south part per week.

Weekly_resampled_data.plot()
```

On the graph, we can see a trend and fluctuations on each day, where we can see that usually north part is more busy, since more 
cycles passes by north part on a weekly basis.

```python
# Now we will plot the line graph that shows the trend how many cycles passes the north and south part on a daily basis.


daily_cycle_passing = data2.resample('D').sum()
daily_cycle_passing .plot(style=[':', '-', '--'])
plt.ylabel('Daily bicycle count')
```

```python
# After the weekly and daily passing, we now see how many cycles passes by particular time period

data2.groupby(data2.index.time).sum().plot()
```

Now create an another dataset which contains the data about finance department and the amount they spend on various departments.

```python
from urllib.request import urlretrieve
URL="https://data.winnipeg.ca/resource/fgvi-j9jp.csv"

urlretrieve(URL,'finance.csv')
```

```python
dt=pd.read_csv('finance.csv')  #read the finance.csv file
```

```python
dt  # print the database
```

```python
dt.head(5)  # return the first 5 rows
```

```python
dt.shape    # returns the number of rows and columns
```

```python
dt.info()
```

```python
# Now we will print the total expected budget decided for each department

department_budget=dt.groupby('department')['_2016_forecast'].sum()
```

```python
department_budget.nlargest(5)  # retuens the top 5 departments with highest budget
```

```python
department_budget.nlargest(5).plot(kind='bar', color='green')
plt.title("Top 5 Department wise expected budget for 2016") 
plt.ylabel("Expected budget")
```

```python
department_service=pd.crosstab(dt["department"], dt["service"], margins=True)
```

```python
department_service  #returns the department and the service it is assigned for work.
```

```python
# returns the department wise service budget for year 2015 for each service

budget_2015=dt.groupby(['department', 'service'])['_2015_adopted'].aggregate('sum').unstack().fillna(0)
```

```python
budget_2015
```

```python
# prints the total budget for each service in year 2015

budget_2015.sum(axis=0)    
```

```python
dt["Total budget for service"]=budget_2015.sum(axis=0).plot(kind='bar',color='orange')
plt.title("Service wise budget in 2015")
```
